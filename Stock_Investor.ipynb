```
import sys
import datetime as date
from datetime import *

class Interest_Calculator():
	def __init__(self, depositDates, depositAmounts, currentDate, currentAmount):
		self.setDepositDates(depositDates, currentDate)
		self.setDepositAmounts(depositAmounts)

		self.currentAmount = currentAmount
		
		self.dailyCompoundedInterestRate = 0.0
		self.annualizedInterestRate = ((1+self.dailyCompoundedInterestRate)**365)-1.0

		self.init_k = 0.0
		self.maxIterations = 100
		self.epsilon = 0.000001
		self.tolerance = 0.0000001
		self.solutionFound = False

	def setDepositDates(self, dates, currentDate):
		self.depositDates = dates
		self.currentDate = currentDate
		self.tDays = [(self.currentDate-date).days for date in self.depositDates]
		#print(self.tDays)

	def setDepositAmounts(self, amounts):
		self.depositAmounts = amounts
		#print(self.depositAmounts)

	def runCalculator(self):
		k = self.init_k
		for i in range(self.maxIterations):
			y, yprime = self.get_funcValues(k)
			#print(k)
			if(abs(yprime) < self.epsilon):
				break

			new_k = k - (y/yprime) #Do Newton's computation
			if(abs(new_k - k) <= self.tolerance): #If the result is within the desired tolerance
				self.solutionFound = True
				break #Done, so leave the loop

			k = new_k #Update k to start the process again

		if (self.solutionFound): #x1 is a solution within tolerance and maximum number of iterations
			self.dailyCompoundedInterestRate =  k
			self.annualizedInterestRate = ((1+self.dailyCompoundedInterestRate)**365)-1.0
		else:
			self.dailyCompoundedInterestRate = 0.0

		return self.solutionFound

	def get_funcValues(self, x):
		func_value = -self.currentAmount;
		diff_func_value = 0.0;
		for i in range(len(self.tDays)):
			func_value += ((1+x)**self.tDays[i])*self.depositAmounts[i]
			diff_func_value += ((1+x)**(self.tDays[i]-1))*self.tDays[i]*self.depositAmounts[i]

		return func_value, diff_func_value

	def getDailyCompoundedInterestRate(self):
		return self.dailyCompoundedInterestRate

	def getAnnualizedInterestRate(self):
		return self.annualizedInterestRate

	def getCurrentInterestRate(self):
		return ((1+self.dailyCompoundedInterestRate)**self.tDays[0])-1.0
```

```
def sipsum(sum,irate,num_months):
  return sum * pow((1+ irate/100),(num_months))

def kittyIn(startsum,startDate,endDate,irate ):
  num = 0
  investmentarray = []
  investDatearray = []
  investDate = startDate
  while True:
    if (investDate - endDate).days > 0:
      break
    investamount = sipsum(startsum,irate,num)
    investmentarray.append(investamount)
    investDate = startDate + relativedelta(months=num)
    investDatearray.append(investDate)
    num +=1
  return investmentarray, investDatearray
```

```
def kittycurrent(investmentarray,investDatearray, withdrawarray,withdrawDatearray,evalDate):
  sum = 0
  for i in range(len(investDatearray)):
    # print(investDatearray[i])
    # print(type(investDatearray[i]))
    if (evalDate - investDatearray[i]).days > 0:
      sum += investmentarray[i]
  # print(withdrawDatearray[i])
  # print(type(withdrawDatearray[i]))
  for i in range(len(withdrawDatearray)):
    if (evalDate - withdrawDatearray[i]).days > 0:
      sum -= withdrawarray[i]
  return sum
```

```

# dfa['dailyPercentChange'] = dfa['Close'].pct_change()
def rollingCandleGenerator(dfa, num):
  dfa[str(num)+'sma'] = dfa['Close'].rolling(window = num).mean().round(2)
  dfa[str(num)+'min'] = dfa['Close'].rolling(window = num).min().round(2)
  dfa[str(num)+'max'] = dfa['Close'].rolling(window = num).max().round(2)
  dfa[str(num)+'start'] = dfa['Close'].shift(num-1).round(2)
  dfa[str(num)+'end'] = dfa['Close'].round(2)
  return dfa

def buynSellLogic(df, numofmonths, buyingThreshold, sellingThreshold):
  num = numofmonths
  dfa = rollingCandleGenerator(df, num)
  dfa[str(num)+'buy'] = dfa['Close']/ dfa[str(num)+'sma'] < buyingThreshold
  dfa[str(num)+'buystrength'] = dfa['Close']/ dfa[str(num)+'sma']

  dfa[str(num)+'sell'] = dfa['Close']/ dfa[str(num)+'sma'] > sellingThreshold
  dfa[str(num)+'sellStrength'] = dfa['Close']/ dfa[str(num)+'sma'] 
  return dfa
```

```
from datetime import date
startsum = 10000
irate = 0
syear = 2014
smonth = 5
sday = 1
eyear = 2018
emonth = 4
eday = 1
evyear = 2020
evmonth = 1
evday = 2 
startDate = date(syear,smonth,sday)
endDate = date(eyear,emonth,eday)
sellingDate = date(evyear,evmonth,evday) 
def StockInvestor(df,startsum=10000,irate = 0, startDate = startDate,endDate = endDate,sellingDate = sellingDate,\
                  avgover =400,buyingThreshold = 2,sellingThreshold = 2 ,percentagesellingatpeak = 0.2, bondrateofInterest = 8, purchasepeak = 0.5):
  stockarray = []
  stockpurchasedDate = []
  investmentarray, investDatearray = kittyIn(startsum,startDate,endDate,irate )
  withdrawarray = []
  withdrawDatearray = []
  #apply logic for buying and selling
  dfa = buynSellLogic(df, avgover, buyingThreshold, sellingThreshold).dropna()
  lasttrans = startDate
  lastinterest = startDate
  for investingDate in dfa.index:
    decDate = investingDate.to_pydatetime().date()
    kittycash = kittycurrent(investmentarray,investDatearray, withdrawarray,withdrawDatearray,decDate)
    #buying
    if kittycash > 0 and dfa.loc[investingDate][str(avgover)+'buy'] and int((decDate - startDate).days) > 0 and int((endDate - decDate ).days) > 0 and int((decDate - lasttrans).days) > 10:
      stockprice = dfa.loc[investingDate]['Close']
      lasttrans = decDate
      numofstockspurchased = math.floor(kittycash / stockprice * purchasepeak)
      withdrawarray.append(numofstockspurchased*stockprice)
      withdrawDatearray.append(decDate)
      stockarray.append(numofstockspurchased)
      stockpurchasedDate.append(decDate)
      # if numofstockspurchased > 0:
      #   print("purchased:",numofstockspurchased,"on",investingDate)
    #selling
    if np.array(stockarray).sum() > 5 and dfa.loc[investingDate][str(avgover)+'sell']  and int((decDate - startDate).days) > 0 and int((endDate - decDate ).days) > 0 and int((decDate - lasttrans).days) > 10 :
      stockprice = dfa.loc[investingDate]['Close']
      lasttrans = decDate
      numberofStockstosell = math.floor(np.array(stockarray).sum() *(percentagesellingatpeak))#int(dfa.loc[investingDate][str(avgover)+'sellStrength'])
      withdrawarray.append(-1 * numberofStockstosell*stockprice)
      withdrawDatearray.append(decDate)
      stockarray.append(-1*numberofStockstosell)
      stockpurchasedDate.append(decDate)
      # print("sold",numberofStockstosell,"on",investingDate)
    #pumping returns from bonds(8 - 10% yearly, every 30 days)
    kittycash = kittycurrent(investmentarray,investDatearray, withdrawarray,withdrawDatearray,decDate)
    kittycash30dayprior = kittycurrent(investmentarray,investDatearray, withdrawarray,withdrawDatearray,decDate - relativedelta(days=30))
    if int((decDate - lastinterest).days) > 30 and kittycash > 1000 and kittycash30dayprior > 1000 and int(( sellingDate- decDate).days) > 0:  
      lastinterest =  decDate
      if kittycash30dayprior > kittycash:
        interestsum = kittycash *bondrateofInterest/1200
      else:
        interestsum = kittycash30dayprior *bondrateofInterest/1200
      # print("invested in bonds:", interestsum)
      withdrawarray.append(-1* interestsum)
      withdrawDatearray.append(decDate)
  #for final evaluation
  for i in range(100):
    try:
      stockprice = dfa.loc[str(sellingDate + relativedelta(days=i))]['Close']
      sellStrength= dfa.loc[str(sellingDate + relativedelta(days=i))][str(avgover)+'sellStrength']
      # if dfa.loc[str(sellingDate + relativedelta(days=i))][str(avgover)+'sellStrength'] > 1.2:
      #   break
    except:  
      pass
  kittycash = kittycurrent(investmentarray,investDatearray, withdrawarray,withdrawDatearray,sellingDate)  
  totalvalue = stockprice * np.array(stockarray).sum() + kittycash  
  calc = Interest_Calculator(investDatearray,investmentarray , sellingDate, totalvalue)
  if(calc.runCalculator()):
    CIannum = calc.getAnnualizedInterestRate()*100
    print("Annualized Compounded Interest = ",CIannum , "%",":",sellStrength)
    print("Interest Until", sellingDate, "is ", calc.getCurrentInterestRate()*100, "%")
  else:
    print("Interest couldn't be calculated. Newton's method failed to converge")
  return totalvalue, sellStrength, CIannum, withdrawarray, withdrawDatearray, investmentarray, investDatearray
```
